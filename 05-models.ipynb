```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
import pandas as pd
```

```python
# create dummy variables from our dataset
pd.get_dummies(tips, drop_first=True)
```

```python
from sklearn import linear_model
```

```python
# 1. create the model object
lr = linear_model.LinearRegression()
```

```python
# 2. fit the model object
lr.fit(X=tips[["total_bill", "size"]], y=tips["tip"])
```

```python
# look at the coefficients
lr.coef_
```

```python
# look at the intercept
lr.intercept_
```

```python
# the model from the model output
# tip = 0.67 + totalbill*0.09 + size*0.19
```

```python
tips_dummy = pd.get_dummies(tips, drop_first=True)[["tip", "total_bill", "smoker_No"]]
```

```python
tips_dummy
```

```python
lr2 = linear_model.LinearRegression()
```

```python
lr2.fit(X=tips_dummy.iloc[:, 1:], y=tips_dummy["tip"])
```

```python
lr2.coef_
```

```python
lr2.intercept_
```

```python
# tip = 0.81 + totalbill*0.10 + smokeing*0.14
```

```python
# fake "new data"
new_data = tips_dummy[["total_bill", "smoker_No"]].tail() # not really new data
```

```python
new_data
```

```python
# use the model to give predicted tip values
new_data["predicted_tips"] = lr2.predict(new_data)
```

```python
new_data
```

```python
# compare fitted model values with "actual" values
# note the predictions are much better than they seem since
# the model didn't actually predict on "new data"
tips_dummy.tail()
```
